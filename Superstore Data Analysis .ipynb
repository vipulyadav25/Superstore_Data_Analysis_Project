# 3.Sum of Sales and Profit by Category 

```python
df.head(3)
```

```python
sales_by_category=df.groupby('Category',as_index=False)['Sales'].sum()
sales_by_category=sales_by_category.sort_values('Sales',ascending=False)
sales_by_category
```

```python
profit_by_category=df.groupby('Category',as_index=False)['Profit'].sum()
profit_by_category=profit_by_category.sort_values('Profit',ascending=False)
profit_by_category
```

```python
profit_margin_by_category=df.groupby('Category',as_index=False)['Profit Margin (%)'].mean()
profit_margin_by_category
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.barplot(x='Sales',y='Category',data=sales_by_category,color='#56DFCF',edgecolor='black')
plt.title('Total Sales by Category',fontweight='bold')
plt.xlabel('Sales in 10^6',fontweight='bold')
plt.ylabel('Category',fontweight='bold')
plt.xticks(rotation=45)
plt.subplot(1,2,2)
sns.barplot(x='Profit',y='Category',data=profit_by_category,color='#00CAFF',edgecolor='black')
plt.title('Total Profit by Category',fontweight='bold')
plt.xlabel('Profit',fontweight='bold')
plt.ylabel('Category',fontweight='bold')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

### Insights:-
#### 1.Most Profitable Category is "Technology" and Least profitable Category is "Furniture"
#### 2.Highest Sales is in "Office Supplies" and Lowest is in "Furniture"
#### Here Sales of Furniture is comparable with the Sales of Office Supplies and Technology but in Profit margin it is very less , So Furniture need to be focused.

# 9.Yearly Profit and Sales Percentage

```python
df.head(2)
```

```python
profit_percentage=df.groupby(df['Order Year'],as_index=False)['Profit'].sum()
total_profit= profit_percentage.Profit.sum()
profit_percentage['Profit (%)']=(profit_percentage.Profit/total_profit)*100
profit_percentage
```

```python
plt.figure(figsize=(10,3))
plt.grid(True)
plt.plot(profit_percentage['Order Year'],profit_percentage['Profit (%)'],'r.-',markersize=10,markeredgecolor='black')
plt.plot(sales_percentage['Order Year'],sales_percentage['Sales (%)'],'g.-',markersize=10,markeredgecolor='black')
plt.title('Yearly Profit and Sales Percentage',fontweight='bold')
plt.ylabel('Profit and Sales in %',fontweight='bold')
plt.xlabel('Order Year',fontweight='bold')    
plt.xticks(rotation=45)
plt.legend(labels=['Profit (%)','Sales (%)'])
plt.show()
```
